```python
import pandas as pd
import numpy as np
#load the dataset as a pandas DataFrame
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
# retrieve numpy array
dataset = df.values
# df=df[0:]
# df.drop()
print(df)
print(df.dtypes)

```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
df.drop(["Department","BusinessTravel","EducationField","JobRole","Over18"],inplace=True,axis=1)
cleanup_nums = {"Gender":     {"Male": 0, "Female": 1},
                "MaritalStatus": {"Single": 0, "Married": 1, "Divorced": 2 },
                "Attrition" : {"Yes" : 1, "No" : 0},
                "OverTime" : {"Yes" : 1, "No" : 0}
                }
```

```python
Y=df["PerformanceRating"]
```

```python

new_obj_df = df.replace(cleanup_nums)
new_obj_df.head()
```

```python
X=new_obj_df
```

```python
new_obj_df.dtypes
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Convert to categorical data by converting data to integers
# X_cat = X.astype(int)

# 10 Features with highest chi-squared statistics are selected
chi2_features = SelectKBest(chi2, k=10)
X_kbest_features = chi2_features.fit_transform(X, Y)

# Reduced Features
print(X.shape[1])
print(X_kbest_features.shape[1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_kbest_features, Y, test_size=0.25,random_state=0)
 
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
Y_pred = classifier.predict(X_test)
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
accuracy_score(Y_test, Y_pred)
```
